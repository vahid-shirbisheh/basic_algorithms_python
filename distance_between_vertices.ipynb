## Building a "not so simple" graph to demonestrate how our algorithm works

```python
g = Graph()
edges = [["H", "X"], ["U", "C"], ["W", "Z"], ["C", "H"], ["X", "V"], 
         ["G", "S"], ["E", "G"], ["J", "H"], ["J", "F"], ["V", "Z"],
         ["X", "E"], ["S", "F"], ["U", "W"], ["A", "F"], ["W", "X"] ]
g.from_edge_list(edges)
# ploting the graph
figure(figsize=(7, 7), dpi=100)
g.plot_connections()
```

## Find the distance of two given vertices in the graph g
Given two vertice v1, v2 in a graph, we show how easy it is to find their distance in linear time using Breadth-First Search implemented in the method "BFS_layers".

```python
def find_dist(graph, v1, v2):
    d = 0
    layers = graph.BFS_layers(v1)
    for i, layer in enumerate(layers):
        if v2 in layer:
            d = i
            break
    return d
```

```python
print("The distance between A and Z = ", find_dist(g, "A", "Z"))
print("The distance between A and X = ", find_dist(g, "A", "X"))
```
