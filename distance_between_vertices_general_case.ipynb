# Distances between vertices of a (possibly disconnected) graph

Here we first decompose the graph into its connected components. Then we find the distance between two given vertices of the graph. 

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import sys
sys.path.append('C:\\Users\\acer\\basic_algorithms_python\\graph')  # use the address of graph class in your PC
from graph import Graph
```

# A disconnected graph
For another example of decomposing a graph into its connected components, see gtest8.py in graph directory.

```python
g = Graph()
edges = [["T", "S"], ["S", "R"], ["U", "T"], ["V", "T"], ["S", "U"],
         ["A", "B"], ["A", "C"], ["C", "D"], ["D", "B"], 
         ["E", "F"], ["G", "E"], ["G", "F"], 
         ["M", "N"],
         ["X", "X"] ]
g.from_edge_list(edges)
# ploting the graph
figure(figsize=(10, 6), dpi=110)
g.plot_connections()
```

# Decomposition into the connected components

```python
components = g.connected_components()
for graph in components:
    print(graph)
    figure(figsize=(6, 4), dpi=90)
    graph.plot_connections()
```

```python
def find_dist(graph, v1, v2):
    d = "infinity"
    for g in components:
        if v1 in g.adj_list.keys() and v2 in g.adj_list.keys():
            layers = g.BFS_layers(v1)
            for i, layer in enumerate(layers):
                if v2 in layer:
                    d = i
                    return d
    return d
```

```python
print("The distance between V and R = ", find_dist(g, "V", "R"))
print("The distance between F and X = ", find_dist(g, "F", "X"))
```
